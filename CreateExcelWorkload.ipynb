```python
import pandas as pd
import openpyxl

# Function to process and aggregate data from the given files
def process_and_aggregate(files):
    aggregated_data = pd.DataFrame()
    for file in files:
        # Load the data
        data = pd.read_csv(file)
        # Adjust the data format: 20 + decimal part
        data.iloc[:, 1:] = data.iloc[:, 1:].applymap(lambda x: x)
        # Rename the first column
        data.rename(columns={data.columns[0]: "Time [min]"}, inplace=True)
        # Append to the aggregated data
        aggregated_data = aggregated_data.append(data)
    return aggregated_data
```

## Packet Loss

```python


folder = "/content/drive/MyDrive/Colab_Notebooks/Thesis/Data/PacketLoss/"
# File paths for the AtoB and BtoA datasets
fixed_files = [folder+'FixedDelay/OWD_Fixed_5min.csv']

normal_files = [folder+'NormalDelay/OWD_Normal_5min.csv']

sin_files = [folder+'SinusoidalDelay/OWD_Sinusoidal_1min.csv']

# Function to process and aggregate data from the given files
def process_and_aggregate(files):
    aggregated_data = pd.DataFrame()
    for file in files:
        # Load the data
        data = pd.read_csv(file)
        # Adjust the data format: 20 + decimal part
        data.iloc[:, 1:] = data.iloc[:, 1:].applymap(lambda x: 20 + x)
        # Rename the first column
        data.rename(columns={data.columns[0]: "Time [min]"}, inplace=True)
        # Append to the aggregated data
        aggregated_data = aggregated_data.append(data)
    return aggregated_data

# Process the data for both AtoB and BtoA
fixed_data = process_and_aggregate(fixed_files)
normal_data = process_and_aggregate(normal_files)
sin_data = process_and_aggregate(sin_files)

# Creating an Excel workbook with two sheets for AtoB and BtoA
excel_writer = pd.ExcelWriter(folder+'PacketLoss_workload.xlsx', engine='openpyxl')

# Writing data to the sheets
fixed_data.to_excel(excel_writer, sheet_name='OWD_AtoB_Fixed', index=False)
normal_data.to_excel(excel_writer, sheet_name='OWD_BtoA_Normal', index=False)
sin_data.to_excel(excel_writer, sheet_name='OWD_BtoA_Sin', index=False)


# Saving the Excel workbook
excel_writer.save()

# Providing the path to the created Excel workbook
excel_workbook_path = excel_writer.path
excel_workbook_path
```
